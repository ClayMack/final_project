```python
# Function to return main style based on substyle 

def get_big_style(x):
    for item in styles:
        if x in item["substyles"]:
            return item["style"]
```

```python
# Apply Function and create new column
df["BA_Big_styles"] = df.beer_style.apply(get_big_style)
```
